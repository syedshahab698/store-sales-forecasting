# Importing Data into dataframe - 

```python
import pandas as pd
import seaborn as sns
# data set 1
holiday_events = pd.read_parquet('{}'.format('holidays_events.parquet.gzip'),)
holiday_events.head()

# data set 2
oil = pd.read_parquet('{}'.format('oil.parquet.gzip'),)
oil.head()

# data set 3
sample_submission = pd.read_parquet('{}'.format('sample_submission.parquet.gzip'),)
sample_submission.head()

# data set 4
stores = pd.read_parquet('{}'.format('stores.parquet.gzip'),)
stores.head()

# data set 5
test = pd.read_parquet('{}'.format('test.parquet.gzip'),)
test.head()

# data set 6
train = pd.read_parquet('{}'.format('train.parquet.gzip'),)
train.head()

# data set 7
transactions = pd.read_parquet('{}'.format('transactions.parquet.gzip'),)
transactions.head()
```

```python
# 1) what are the columns for each data set?
data_1 = list(holiday_events.columns)
data_2 = list(oil.columns)
data_3 = list(sample_submission.columns)
data_4 = list(stores.columns)
data_5 = list(test.columns)
data_6 = list(train.columns)
data_7 = list(transactions.columns)

print(data_1)
print(data_2)
print(data_3)
print(data_4)
print(data_5)
print(data_6)
print(data_7)
```

```python
# Whether column is numeric or categorical for each data set?
# My analysis is based on the variation of values in the column.
print("Holiday Events Dataset - 1")
for col in list(holiday_events.columns):
    if holiday_events[col].nunique() > 50:
        print(f'{col} --> numerical')
    else:
        print(f'{col} --> catrgorical')
        
print("\n")
      
print("Oil Dataset - 2")
for col in list(oil.columns):
    if oil[col].nunique() > 50:
        print(f'{col} --> numerical')
    else:
        print(f'{col} --> catrgorical')
        
print("\n")

print("Sample_submission Dataset - 3")
for col in list(sample_submission.columns):
    if sample_submission[col].nunique() > 50:
        print(f'{col} --> numerical')
    else:
        print(f'{col} --> catrgorical')  
        
print("\n")
print("Stores Dataset - 4")
for col in list(stores.columns):
    if stores[col].nunique() > 50:
        print(f'{col} --> numerical')
    else:
        print(f'{col} --> catrgorical')
        
print("\n") 
print("Test Dataset - 5")
for col in list(test.columns):
    if test[col].nunique() > 50:
        print(f'{col} --> numerical')
    else:
        print(f'{col} --> catrgorical')
        
print("\n")
print("Train Dataset - 6")
for col in list(train.columns):
    if train[col].nunique() > 50:
        print(f'{col} --> numerical')
    else:
        print(f'{col} --> catrgorical')
        
print("\n")
print("Transactions Dataset - 7")
for col in list(transactions.columns):
    if transactions[col].nunique() > 50:
        print(f'{col} --> numerical')
    else:
        print(f'{col} --> catrgorical')
```
